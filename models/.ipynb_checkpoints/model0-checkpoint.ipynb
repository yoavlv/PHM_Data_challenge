```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import scipy.fft
from scipy.stats import skew, kurtosis
import scipy.signal
import pywt
from scipy.stats import skew, kurtosis
# from PyEMD import EMD
import seaborn as sns 
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from scipy.io import loadmat
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout
import matplotlib.pyplot as plt
from keras.layers import RepeatVector
```

```python
# features_df['Tachometer_FFT'] = features_df['Tachometer'].apply(apply_fft)

# def apply_fft(series):
#     # Applying Fourier Transform
#     freq_domain = fft(series)
#     return np.abs(freq_domain)

def feature_extraction(df):
    features = {}
    # 1. Statistical Features
    
    for column in df.columns:
        features[f'{column}_mean'] = df[column].mean()
        features[f'{column}_median'] = df[column].median()
        features[f'{column}_std'] = df[column].std()
        features[f'{column}_var'] = df[column].var()
        features[f'{column}_range'] = df[column].max() - df[column].min()
        features[f'{column}_skew'] = skew(df[column])
        features[f'{column}_kurtosis'] = kurtosis(df[column])

    # 2. Time Series Features
    for column in df.columns:
        features[f'{column}_autocorr_lag1'] = df[column].autocorr(lag=1)

    # 3. Frequency Domain Features
    for column in df.columns:
        column_data = df[column].to_numpy()
        freq_data = scipy.fft.fft(column_data)
        freq_amplitudes = np.abs(freq_data)
        features[f'{column}_dominant_freq'] = np.argmax(freq_amplitudes)
        features[f'{column}_spectral_entropy'] = -np.sum((freq_amplitudes/np.sum(freq_amplitudes)) * np.log(freq_amplitudes/np.sum(freq_amplitudes)))

    # Wavelet Transform Features
    for column in df.columns:
        column_data = df[column].to_numpy()
        coeffs = pywt.wavedec(column_data, 'db1')
        for i, coeff in enumerate(coeffs):
            features[f'{column}_wavelet_avg_coeff_{i}'] = np.mean(coeff)
            features[f'{column}_wavelet_std_coeff_{i}'] = np.std(coeff)

    # 4. Cross-Column Features
    features['input_output_corr'] = df['Input Voltage'].corr(df['Output Voltage'])
    features['input_tachometer_corr'] = df['Input Voltage'].corr(df['Tachometer'])
    features['output_tachometer_corr'] = df['Output Voltage'].corr(df['Tachometer'])

    feature_df = pd.DataFrame([features])
    
    return feature_df

def create_features_df():
    all_features_df = pd.DataFrame()
    for i in range(1, 561):
        columns = ['Input Voltage', 'Output Voltage', 'Tachometer']
        temp_df = pd.read_csv(f'PHM09_competition_1/Run_{i}.csv', names=columns)
        res_df = feature_extraction(temp_df)
        
        res_df['run_id'] = i
        reordered_columns = ['run_id'] + [col for col in res_df.columns if col != 'run_id']
        res_df = res_df[reordered_columns]
        
        all_features_df = pd.concat([all_features_df, res_df], ignore_index=True)
    return all_features_df

features_df = create_features_df()
#  התפלגות נורמלית = התפלגאות גאוסנית
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro

def plot_distributions(df):
    df = df.drop(columns =['run_id'])
    dist = {}
    for column in df.select_dtypes(include=[np.number]).columns:
        # Check if the column has more than one unique value
        try:
            if len(df[column].dropna().unique()) > 1:
                # Shapiro-Wilk Test for Normality
                # add ln 2 functio
                stat, p = shapiro(df[column].dropna())
                alpha = 0.05
                distribution_type = 'Normal' if p > alpha else 'Not Normal'
                
                # Convert the column data to numpy array for plotting
                column_data = df[column].dropna().values

                # Plotting the distribution
#                 plt.figure(figsize=(6, 4))
#                 sns.histplot(column_data, kde=True)
#                 plt.title(f'{column} - {distribution_type} Distribution (p={p:.3f})')
#                 plt.xlabel(column)
#                 plt.ylabel('Frequency')
#                 plt.show()
                dist[column] = distribution_type
                print(f'{column} - {distribution_type} Distribution (p={p:.3f})')
            else:
                print(f"{column}: Only one unique value, distribution plot not applicable.")
        except:
            pass

    return dist 
# Example usage
# plot_distributions(your_dataframe)
dist = plot_distributions(features_df)
```

```python
def top_correlated_features(df, target_column, x):
    correlation_matrix = df.corr()
    correlations = correlation_matrix[target_column].drop(target_column).abs()
    sorted_correlations = correlations.sort_values(ascending=False)
    top_features = sorted_correlations.head(x)
    return top_features.to_dict()

top_features_dict = top_correlated_features(features_df, 'Tachometer_mean', 10)
top_features_dict

df_corr = features_df[list(top_features_dict.keys())]
```

```python
df_corr
```

```python
X = np.array(df_corr).reshape((df_corr.shape[0], 1, df_corr.shape[1]))

# Building LSTM Autoencoder
model = Sequential()
# Encoder
model.add(LSTM(50, activation='relu', input_shape=(1, X.shape[2]), return_sequences=True))
model.add(LSTM(30, activation='relu', return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(X.shape[2]))

# Decoder
model.add(RepeatVector(1))
model.add(LSTM(30, activation='relu', return_sequences=True))
model.add(LSTM(50, activation='relu', return_sequences=True))
model.add(Dropout(0.2))
model.add(Dense(X.shape[2]))

model.compile(optimizer='adam', loss='mse')

# Training LSTM Autoencoder
model.fit(X, X, epochs=10, batch_size=32)
```

```python
predicted = model.predict(X)

# Ensure that X and predicted are reshaped to 2D arrays if they are not already
X_reshaped = X.reshape(X.shape[0], -1)  # Reshaping to a 2D array with runs as rows
predicted_reshaped = predicted.reshape(predicted.shape[0], -1)

# Calculate the mean squared error for each run
mse = np.mean(np.power(X_reshaped - predicted_reshaped, 2), axis=1)

# Ensure mse is a 1D array
mse = mse.flatten()

# Check the length of mse

# Proceed with the rest of your anomaly detection
threshold = np.quantile(mse, 0.9)
anomalies = mse > threshold

# Ensure anomalies is a 1D array
anomalies = anomalies.flatten()

# Creating a results dataframe
results_df = pd.DataFrame(anomalies.astype(int), columns=['Fault Prediction'])
results_df.index.name = 'Run Number'
results_df.index += 1
```

```python
from scipy.fftpack import fft

# Example feature: FFT of the input voltage
def extract_fft_features(signal):
    N = len(signal)
    fft_values = fft(signal)
    fft_norm = np.abs(fft_values) / N  # Normalize
    return fft_norm[:N // 2]  # Return only the positive frequencies

# Extract features for each sample
features = np.array([extract_fft_features(row) for row in data_scaled])
```

```python
from sklearn.ensemble import IsolationForest

# Initialize and fit the model
model = IsolationForest(contamination='auto')  # contamination: the proportion of outliers in the data set
model.fit(features)

# Anomaly scores (the lower, the more anomalous)
scores = model.decision_function(features)

# Detect anomalies (True for normal, False for anomaly)
anomalies = model.predict(features) < 0
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.scatter(range(len(scores)), scores, c=anomalies, cmap='coolwarm')
plt.xlabel('Data Point')
plt.ylabel('Anomaly Score')
plt.title('Anomaly Detection')
plt.show()
```
